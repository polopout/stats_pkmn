```python
import pandas as pd
from unidecode import unidecode
from collections import Counter

# Exemple de colonne
# df["pokemon_favoris"] = ["galeking-libegon-aligatueur-dracolosse-bastiodon", "ho-oh-pikachu-dracolosse"]

def extraire_pokemons(val):
    if pd.isna(val):
        return []
    val = str(val).lower()
    val = unidecode(val)       # retire accents
    val = val.replace(" ", "")  # retire espaces
    # On protège ho-oh en le remplaçant temporairement
    val = val.replace("ho-oh", "hoohtemp")
    # Split par tiret
    pokemons = val.split("-")
    # On remet ho-oh
    pokemons = ["ho-oh" if p=="hoohtemp" else p for p in pokemons]
    return pokemons

# Appliquer et aplatir la liste
all_pokemons = df["pokemon_favoris"].apply(extraire_pokemons)
flat_list = [p for sublist in all_pokemons for p in sublist]

# Compter
pokemon_counts = Counter(flat_list)

# Top 5
top5 = pokemon_counts.most_common(10)
print("Top 5 des Pokémon favoris :")
for p, count in top5:
    print(f"{p} : {count}")
```
